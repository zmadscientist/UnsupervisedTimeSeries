```python
import cv2
import numpy as np

def extract_coordinates(image_path):
    # Load the image
    image = cv2.imread(image_path)

    # Define the color range for the red line
    lower_red = np.array([0, 0, 100])
    upper_red = np.array([50, 50, 255])

    # Create a mask for the red line
    mask = cv2.inRange(image, lower_red, upper_red)

    # Find the coordinates of the red line
    coordinates = cv2.findNonZero(mask)

    # Convert the coordinates to a list of (x, y) tuples
    coordinates = [(pt[0][0], pt[0][1]) for pt in coordinates]

    return coordinates

# Use the function
coordinates = extract_coordinates('TorontoRed.png')
print(coordinates)
```

```python
from PIL import Image

# Load the image
image = Image.open('TorontoRed.png')

# Convert the image to RGB if it's not
image = image.convert('RGB')

# Initialize a list to store the coordinates of red pixels
red_pixels = []

# Iterate over each pixel in the image
for x in range(image.width):
    for y in range(image.height):
        # Get the RGB color of the pixel
        r, g, b = image.getpixel((x, y))
        # Check if the pixel is red (assuming pure red color)
        if r > 140 and g < 100 and b < 100:
            # Add the pixel's coordinates to the list
            red_pixels.append((x, y))

# Output the list of red pixel coordinates
red = np.array(red_pixels)
red[:,1] = 255 - red[:,1]
```

```python
import matplotlib.pyplot as plt
plt.scatter(red[:,0], red[:,1], c='r',s=1, alpha = .3)
plt.grid()
```
